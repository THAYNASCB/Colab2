```python
#numero identificador unico para cada usuario, iniciando com valor 1, v.02.1.
identificadorUsuarios = 1

#inicializando a lista de contatos vazia.
lista = [] 

def backupContatos():
  with open('contatos.bkup','w') as arquivo:
    arquivo.write('Número,Nome,Email,Telefone,Twitter,Instagram')
    for registro in lista:
      arquivo.write(str(registro['id']) + ',' + registro['nome'] + ',' + registro['tel'] + ',' + registro['email'] + ',' + registro['twitter'] + ',' + registro['instagram'])
    arquivo.close()
    print('backup salvo com sucesso')

def gerarRelatorio() :
  print('Número \t Nome \t\t\t Email \t\t\t Telefone \t\t\t Twitter \t\t\t Instagram');
  for registro in lista:
    print(str(registro['id']) + '\t' + registro['nome'] + '\t' + registro['tel'] + '\n' + registro['email'] + '\t' + registro['twitter'] + '\t' + registro['instagram'])

#função para impedir cadastro duplicado.
def existe_contato(nome):
  if len(lista) > 0:
    for contato in lista:
      if contato['nome'] == nome:
        return True
  return False

#função que adiciona contato.
def adicionarUsuarios(num):
  while(num > 0):
    email = input("Digite e-mail do contato:")
    nome = input("Digite o nome:")
    telefone = input("Digite o telefone:")
    twitter = input("twitter:")
    instagram = input("instagram:")

#função para acusar erro de contato já cadastrado.
    if existe_contato(nome): 
      print ("Erro: Esse nome já está cadastrado.")
      return

# acessando a variável global definida no começo do programa.   
    global identificadorUsuarios

#função dados contato incluindo "palavra chave" como nome único de contato.
    contato = {
    "id": identificadorUsuarios,
    "email":email,
    "nome":nome,
    "tel": telefone,
    "twitter": twitter,
    "instagram": instagram
    }
    lista.append(contato)
    print("O contato {} foi cadastrado com sucesso\n".format(contato['nome']))
    num -= 1
    identificadorUsuarios += 1

#função que adiciona contato.
def adicionar():
  email = input("Digite e-mail do contato:")
  nome = input("Digite o nome:")
  telefone = input("Digite o telefone:")
  twitter = input("twitter:")
  instagram = input("instagram:")

 #função que acusa se o contato já está na lista 
  if existe_contato(nome): 
    print ("Erro: Esse nome já está cadastrado.")
    return
  
# acessando a variável global definida no começo do programa.   
  global identificadorUsuarios

 #função que contém palavra nome único do contato.
  contato = {
  "id": identificadorUsuarios,
  "email":email,
  "nome":nome,
  "tel": telefone,
  "twitter": twitter,
  "instagram": instagram
  }
  lista.append(contato)
  identificadorUsuarios += 1
  print("O contato {} foi cadastrado com sucesso\n".format(contato['nome']))

#função que altera o contato.
def alterar(nome):
  if existe_contato(nome):
    email = input("Digite o email:")
    telefone = input("Digite o telefone:")
    twitter = input("twitter:")
    instagram = input("instagram:")
    indice: int = 0
    for contato in lista:
       if contato['nome']==nome:
         break;
       indice = indice + 1
    lista[indice]['email'] = email
    lista[indice]['tel'] = telefone
    lista[indice]['twitter'] = twitter
    lista[indice]['instagram'] = instagram
    print("O contato {} foi alterado com sucesso\n".format(nome))
  else:
    print("O contato {} não foi encontrado\n".format(nome))

#função que exclui dados do contato.
def excluir(nome):
  if existe_contato(nome): 
    indice = 0
    for contato in lista:
       if contato['nome']==nome:
         break;
       indice = indice + 1
    del lista[indice]
    print("O contato {} foi excluido com sucesso\n".format(nome))
  else:
    print("O contato {} não foi encontrado\n".format(nome))

#função para buscar contato pela "chave" nome.  
def buscarPorNome(nome):
  if len(lista) > 0:
    for contato in lista:
       if contato['nome']==nome:
          print(contato['id'])
          print(contato['email'])
          print(contato['tel'])
          print(contato['twitter'])
          print(contato['instagram'])
          for contato in lista:
            if contato['nome']==nome:
             break;
          print("O contato {} foi encontrado com sucesso\n".format(nome))
       else:
          print("O contato {} não foi encontrado\n".format(nome))

#função para listar dados do contato.
def listar():
  for contato in lista:
    print(contato['id'])
    print(contato['email'])
    print(contato['nome'])
    print(contato['tel'])
    print(contato['twitter'])
    print(contato['instagram'])

 #função para chamar o menu de geração de dados do contato.
def menu():
  while True:
    print("===Agenda Telefônica===")
    print("1 - Adicionar contato")
    print("2 - Alterar contato")
    print("3 - Excluir contato")
    print("4 - Buscar contato por nome4")
    print("5 - Listar contatos")
    print("6 - Incluir mais de um contato ao mesmo tempo")
    print("7 - Gerar Relatório")
    print("8 - Fazer Backup dos Contatos")
    print("9 - Sair")
    opcao = int (input(">"))
    if opcao == 1:
      adicionar()
    elif opcao == 2:
      nome = input("Digite nome do contato que deseja alterar:")
      alterar(nome)
    elif opcao == 3:
      nome = input("Digite nome do contato que deseja excluir:")
      excluir(nome)
    elif opcao == 4:
      nome = input("Digite nome do contato que deseja buscar:")
      buscarPorNome(nome)
    elif opcao == 5:
      listar()
    elif opcao == 6:
      num = int (input("Digite o número de usuários a serem cadastrados:"))
      adicionarUsuarios(num)
    elif opcao == 7:  
      gerarRelatorio()
    elif opcao == 8:  
      backupContatos()
    elif opcao == 9:
      print("Saindo do programa...")
      break
    else:
      print("Opção inválida. Por favor, tente novamente!") 

#função para chamar o menu.
menu()
```
